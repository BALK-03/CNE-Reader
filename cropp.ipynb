```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = r"C:\Users\HP\Downloads\cartez.jpg"  
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.title('Original Image')
plt.show()
```

```python


# Define the coordinates of the region you want to crop 
x1, y1 = 195, 254  
x2, y2 = 569, 347 

# Crop the region of interest from the image
cropped_region = image[y1:y2, x1:x2]

# Display the cropped region in the output section
plt.imshow(cv2.cvtColor(cropped_region, cv2.COLOR_BGR2RGB))
plt.axis('off')  
plt.show()
```
